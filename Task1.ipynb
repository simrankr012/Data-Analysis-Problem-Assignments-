```python
import pandas as pd
import csv
import numpy as np
```

```python
lines = np.genfromtxt("sheet2.csv", delimiter=",", dtype=None,skip_header=1)
my_dict = dict()

for line in lines:
    if line[0] in my_dict:
        # append the new number to the existing array at this slot
        my_dict[line[0]].append(line[1])
    else:
        # create a new array in this slot
        my_dict[line[0]] = [line[1]]

for i in my_dict:
    print i, my_dict[i]
        
        
```
